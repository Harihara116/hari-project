```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\Hariharan\\Downloads\\pandas\\besent assisement test Netflix+Dataset2021.csv")
```

```python
# movie released in the year 2000
movies_2000 = df[(df['Category'] == 'Movie') & (df['Release_Date'] == 2000)]
```

```python
#the title of tv shows that were released only in india
Country=["India"]
```

```python
df[df.Country.isin(Country)]
```

```python
#the Top 10 directors who gave the highest number of 
TV shows and movies to Netflix
df4=df['Director'].value_counts().head(10)
```

```python
df4
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df4.index, y=df4.values)
plt.title('Top 10 Directors')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```

```python
#category is movie and type is comedies” or “county is united kingdom
df[((df['Category'] == "Movie") & (df['Type'] == "Comedies") & (df['Country'] == "United Kingdom"))]
```

```python
#movies/shows, Tom Cruise were cast
df[((df['Cast'] == "Tom Cruise"))]
```

```python
#all Ratings defined by Netflix
df['Rating'].unique()
```

```python
#content ratings in a graph
plt.figure(figsize=(10, 6))
sns.countplot(x='Rating', data=df)
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# the ratings and no. of unique ratings
len(df['Rating'].unique())
```

```python
#the highest no. of .Tv Shows
tv_shows_by_country = df[df['Category'] == 'TV Show'].Country.value_counts()
```

```python
movies_by_country = df[df['Category'] == 'Movie'].Country.value_counts()
```

```python
tv_shows_by_country.max()
```

```python
#all the countries and no .of tv shows in a graph
plt.figure(figsize=(10, 6))
tv_shows_by_country.head(10).plot(kind='bar')
plt.title('TV Shows by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
#each countries tv show and movie count and show them in a graph
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
tv_shows_by_country.head(10).plot(kind='bar')
plt.title('TV Shows by Country')
plt.xlabel('Country')
plt.ylabel('Count')
```

```python
plt.subplot(1, 2, 2)
movies_by_country.head(10).plot(kind='bar')
plt.title('Movies by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
compare_df = pd.DataFrame({'TV Shows': tv_shows_by_country, 'Movies': movies_by_country})
compare_df.head(10).plot(kind='bar')
plt.title('TV Shows vs Movies by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()

```

```python
# Sort the dataset by year
netflix_data_sorted = df.sort_values(by='Release_Date')
```

```python
netflix_data_sorted
```

```python
year_count = netflix_data_sorted['Release_Date'].value_counts().sort_index()
```

```python
# the count of movies/tv shows released in a year in line chart
plt.figure(figsize=(10, 6))
year_count.plot(kind='line')
plt.title('Movies/TV Shows Released by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
#Top 5 directors in the Netflix based on the count of movies/tv shows directed 
top_directors = netflix_data_sorted['Director'].value_counts().head(5)
```

```python
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar')
plt.title('Top 5 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Count')
plt.show()
```
